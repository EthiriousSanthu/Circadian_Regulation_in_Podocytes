### TPM

```r
gene_length_extract <- function(data){
  gene_symbols <- rownames(data)
  gene_map <- select(org.Hs.eg.db, keys = gene_symbols, keytype = "SYMBOL", columns = "ENSEMBL")
  gene_map <- gene_map[!is.na(gene_map$ENSEMBL), ]
  gene_map <- gene_map[!duplicated(gene_map$SYMBOL), ]

  gene_length_gc <- getGeneLengthAndGCContent(org = "hg38", mode = "org.db", id = gene_map$ENSEMBL)
  gene_length_gc <- as.data.frame(gene_length_gc)
  gene_length_gc$gName <- gene_map$SYMBOL
  gene_length_gc <- aggregate(.~gName, data=gene_length_gc, FUN=mean)

  return(gene_length_gc)
}
```

```r
counts_to_tpm <- function(counts, featureLength, meanFragmentLength) {
 
  # Ensure valid arguments.
  stopifnot(length(featureLength) == nrow(counts))
  stopifnot(length(meanFragmentLength) == ncol(counts))
 
  # Compute effective lengths of features in each library.
  effLen <- do.call(cbind, lapply(1:ncol(counts), function(i) {
    featureLength - meanFragmentLength[i] + 1
  }))
 
  # Exclude genes with length less than the mean fragment length.
  idx <- apply(effLen, 1, function(x) min(x) > 1)
  counts <- counts[idx,]
  effLen <- effLen[idx,]
  featureLength <- featureLength[idx]
 
  # Process one column at a time.
  tpm <- do.call(cbind, lapply(1:ncol(counts), function(i) {
    rate = log(counts[,i]) - log(effLen[,i])
    denom = log(sum(exp(rate)))
    exp(rate - denom + log(1e6))
  }))
 
  # Copy the row and column names from the original matrix.
  colnames(tpm) <- colnames(counts)
  rownames(tpm) <- rownames(counts)[idx]
  return(tpm)
}
```

## Using Podocyte Circadian Genes

```r
CRD_score_cal <- function(rds_file_path, circadian_genes) {
  data <- readRDS(rds_file_path)

  exp_mat <- GetAssayData(data, assay = "Xenium", slot = "counts")
  podocytes <- WhichCells(data, expression = predicted.annotation.l2 == "Podocyte")

  pod_exp <- exp_mat[, podocytes]
  pod_exp_df <- as.data.frame(pod_exp)

  gene_len <- gene_length_extract(data)

  pod_fin <- pod_exp_df[rownames(pod_exp_df) %in% gene_len$gName[!is.na(gene_len$length)], ]

  pod_tpm <- counts_to_tpm(pod_fin,
                           featureLength = gene_len$length[match(rownames(pod_fin), gene_len$gName)],
                           meanFragmentLength = rep(1, ncol(pod_fin)))
  
  crd <- cal_CRDscore(expr = pod_tpm, circadians = circadian_genes, n.bins = 50, study.type = "scRNAseq")

  return(crd)
}
```

```r
folder_path <- "RDS_Files"

rds_files <- list.files(folder_path, pattern = "\\.RDS$", full.names = TRUE)
```

```r
Xsamples <- str_replace(basename(rds_files), "\\.RDS$", "")
```

```r
pod_data <- readxl::read_xlsx("Podocyte circadian transcriptomics meta2d BHQ _0.1 .xlsx", col_names = TRUE)
```

```r
pod_cir_genes <- pod_data$Gene_Symbol[pod_data$JTK_BH.Q < 0.01]
pod_cir_genes <- toupper(pod_cir_genes)
```

```r
pod_crd_score <- list()
```

```r
for (i in seq_along(rds_files)) {
  cat("Processing", Xsamples[i], "\n")
  crd_score <- CRD_score_cal(rds_files[i], pod_cir_genes)

  pod_crd_score[[Xsamples[i]]] <- crd_score
}
```

### Correlation Study of Podocyte CRD vs PDS

```r
pds_score <- readRDS("PDS.perCell.RDS")
```

```r
extract_sample_id <- function(crd) {
  stringr::str_extract(crd, "PID\\.\\d{2}-\\d{5}\\.\\.rep\\d")
}
```

```r
pod_plot_data <- list()
```

```r
for (pod_crd_sample in names(pod_crd_score)) {
  sample_id <- extract_sample_id(pod_crd_sample)

  crd_vec <- pod_crd_score[[pod_crd_sample]]
  pds_vec <- pds_score[[sample_id]]

  common_cells <- intersect(names(crd_vec), names(pds_vec))

  if (length(common_cells) > 10) {
    df <- data.frame(
      Sample = sample_id,
      Cell = common_cells,
      CRD = crd_vec[common_cells],
      PDS = pds_vec[common_cells]
    )
    pod_plot_data[[sample_id]] <- df
  }
}
```

```r
final_pod_df <- bind_rows(pod_plot_data)
```

```r
filtered_df <- final_pod_df %>% filter(CRD > -500)
```

```r
ggplot(filtered_df, aes(x = PDS, y = CRD, color = Sample)) +
  geom_point(alpha = 0.5) +
  geom_smooth(method = "lm", color = "darkred", se = TRUE) +
  stat_cor(aes(group = 1), method = "spearman") +
  labs(title = "CRD vs PDS for Podocyte Circadian Genes", x = "PDS", y = "CRD") +
  theme_minimal() +
  theme(plot.title = element_text(hjust = 0.5))
```

```r
avg_scr <- final_pod_df %>% 
  group_by(Sample) %>% 
  summarise(
    mean_crd = mean(CRD, na.rm = TRUE),
    sd_crd = sd(CRD, na.rm = TRUE),
    mean_pds = mean(PDS, na.rm = TRUE),
    sd_pds = sd(PDS, na.rm = TRUE),
    spearman_r = cor(PDS, CRD, method = "spearman"),
    p_value = cor.test(PDS, CRD, method = "spearman")$p.value,
    n_cells = n()
  )
```

```r
avg_scr <- avg_scr %>% mutate(Sample_ID = str_extract(Sample, "\\d{2}-\\d{5}"))
```

```r
metadata = readxl::read_xlsx("metadata.xlsx", col_names = TRUE, 3)
```

```r
colnames(metadata)[colnames(metadata) == "ForMe_ID"] <- "Sample_ID"
colnames(metadata)[colnames(metadata) == "Age at Diagnosis"] <- "Age"
colnames(metadata)[colnames(metadata) == "eGFR at biopsy (ml/min, FAS)"] <- "eGFR"
colnames(metadata)[colnames(metadata) == "ACR at biopsy (mg/g)"] <- "ACR"
```

```r
merge_df <- left_join(avg_scr, metadata, by = "Sample_ID")
```

```r
cor.test(merge_df$mean_crd, merge_df$Age, method = "spearman")

ggplot(merge_df, aes(x = Age, y = mean_crd)) +
  geom_point(size = 3, color = 'steelblue') +
  geom_smooth(method = "lm", se = TRUE, color = 'darkred') +
  stat_cor(method = "spearman") +
  labs(title = "Correlation of Avg CRD with Age", x = "Age", y = "Avg CRD") +
  theme_minimal() +
  theme(plot.title = element_text(hjust = 0.5))
```

```r
cor.test(merge_df$mean_crd, merge_df$eGFR, method = "spearman")

ggplot(merge_df, aes(x = eGFR, y = mean_crd)) +
  geom_point(size = 3, color = 'steelblue') +
  geom_smooth(method = "lm", se = TRUE, color = 'darkred') +
  stat_cor(method = "spearman") +
  labs(title = "Correlation of Avg CRD with eGFR", x = "eGFR", y = "Avg CRD") +
  theme_minimal() +
  theme(plot.title = element_text(hjust = 0.5))
```

```r
cor.test(merge_df$mean_crd, merge_df$ACR, method = "spearman")

ggplot(merge_df, aes(x = ACR, y = mean_crd)) +
  geom_point(size = 3, color = 'steelblue') +
  geom_smooth(method = "lm", se = TRUE, color = 'darkred') +
  stat_cor(method = "spearman") +
  labs(title = "Correlation of Avg CRD with ACR", x = "ACR", y = "Avg CRD") +
  theme_minimal() +
  theme(plot.title = element_text(hjust = 0.5))
```
